```python
import geopandas as gpd
import os
import pandas as pd
import folium
import pickle
```

## Create combined attributes data frames

```python
path = 'attributes\\'
```

```python
attributes_1_paths = []

for data_set in list(os.walk(path))[1:]:
    current_attributes_1 = data_set[0] + "\\" + data_set[2][0]
    attributes_1_paths.append(current_attributes_1)
```

```python
attributes_2_paths = []

for data_set in list(os.walk(path))[1:]:
    current_attributes_2 = data_set[0] + "\\" + data_set[2][2]
    attributes_2_paths.append(current_attributes_2)
```

```python
attributes_1_paths
```

```python
attributes_2_paths
```

```python
attributes_1_df_list = []

for i in attributes_1_paths:
    attributes_1_df_list.append(pd.read_csv(i))

attributes_1 = pd.concat(attributes_1_df_list, axis=0)
```

```python
attributes_2_df_list = []

for i in attributes_2_paths:
    attributes_2_df_list.append(pd.read_csv(i))

attributes_2 = pd.concat(attributes_2_df_list, axis=0)
```

```python
attributes_1.reset_index(inplace=True, drop=True)
attributes_1
```

```python
attributes_2.reset_index(inplace=True, drop=True)
attributes_2
```

```python
with open('leadtime_3_naive_performance.pkl', 'rb') as f:
    leadtime_3_naive_performance = pickle.load(f)

leadtime_3_naive_performance['gauge_id'] = leadtime_3_naive_performance.index
leadtime_3_naive_performance
```

## Make visualizations

```python
attributes_1_names = ["p_mean",
                      "pet_mean",
                      "aridity",
                      "frac_snow",
                      "moisture_index",
                      "seasonality",
                      "high_prec_freq",
                      "high_prec_dur",
                      "low_prec_freq",
                      "low_prec_dur"]
```

For the following cells to work, make sure you unzip `combined.zip` which contains `combined.geojson`.

The cells below are very computationally demanding:

```python
m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data='combined.geojson',
    name="choropleth",
    data=leadtime_3_naive_performance,
    columns=["gauge_id", "KGE"],
    key_on="feature.properties.gauge_id",
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="KGE",
).add_to(m)

folium.LayerControl().add_to(m)

m
```

```python
m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data='combined.geojson',
    name="choropleth",
    data=leadtime_3_naive_performance,
    columns=["gauge_id", "NSE"],
    key_on="feature.properties.gauge_id",
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Naive NSE",
).add_to(m)

folium.LayerControl().add_to(m)

m
```
